# 마이닝 알고리즘

## 공분산

2개의 확률변수의 선형관계를 나타내는 값

상관이 있다면, 양의 공분산을 가진다.

- 두 변수의 관계를 수치적으로 나타내는 통계적 측정값
- 숫자가 높을 수록 강한 관계

## 상관계수

- 공분산 에서 스케일을 제거한 계수
- 두 변수 사이의 강도와 방향을 보여준다.
