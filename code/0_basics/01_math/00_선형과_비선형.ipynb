```python
import numpy as np
import matplotlib.pyplot as plt

# 선형 함수 정의
def linear_model(x, weight=2, bias=3):
    return weight * x + bias

# 비선형 함수 정의 sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))   # np.exp(-x) = e의 -x 제곱  e 자연상수(= 2.71828...)

np.exp(2)
np.exp(-2)
```

```python
# 테스트용 데이터 생성 range of values from -10 to 10
inputs = np.linspace(-10, 10, 100)
inputs
```

```python
# 선형 함수 적용
linear_outputs = linear_model(inputs)
linear_outputs
```

```python
# 선형함수의  output을 비선형 함수의 input으로 적용
# np.set_printoptions(precision=2)
non_linear_outputs = sigmoid(linear_outputs)
np.round(non_linear_outputs, 3)
```

```python
# 시각화
plt.figure(figsize=(12, 5))

# Linear plot
plt.subplot(1, 2, 1)
plt.plot(inputs, linear_outputs, label='Linear Model Output')
plt.title('Linear Model')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
```

```python
# Non-linear plot
plt.subplot(1, 2, 2)
plt.plot(inputs, non_linear_outputs, label='Non-linear Model Output')
plt.title('Non-linear Transformation with Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()

plt.tight_layout()
plt.show()
```
