### or 연산자
|좌측|우측|결과|

|True|True|True|

- and: 콜라랑 사이다 사와 --> 콜라, 사이다 둘 다 사가야 혼나지 않음

````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

````python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

## elif 구문
- 딱 두경우로 구분이 안되는 경우
- 계절을 출력했던 프로그램을 생각해보자. (4경우)
```
if 조건A:
    조건A가 참일 때 실행할 문장
elif 조건B:
    조건A가 거짓이고, 조건B가 참일 때 실행할 문장
elif 조건C:
    조건A와 조건B가 거짓이고, 조건C가 참일 때 실행할 문장
...
else:
    위의 조건이 모두 거짓일 때 실행할 문장
```


````python
if True:
  print('if True')
elif True:
  print('elif True')
else:
  print('False')
````

````python
# 계절을 구분하는 프로그램을 만들어보자
now = datetime.datetime.now()
month = now.month

if 3 <= month <= 5:
  print('이번 달은 {}월로 봄입니다.'.format(month))

elif 6 <= month <= 8:
  print('이번 달은 {}월로 여름입니다.'.format(month))

elif 9 <= month <= 11:
  print('이번 달은 {}월로 가을입니다.'.format(month))

else:
  print('이번 달은 {}월로 겨울입니다.'.format(month))
````
