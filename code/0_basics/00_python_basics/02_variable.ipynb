복합 대입 연산자의 종류

|연산자 이름|설명|
|-|-|
|+=|숫자 덧셈 후 대입|
|-=|숫자 뺄셈 후 대입|
|*=|숫자 곱셈 후 대임|
|/=|숫자 나눗셈 후 대입|
|%=|숫자의 나머지를 구한 후 대입|
|//=|숫자의 몫을 구한 후 대입|
|**=|숫자 제곱 후 대입|

```python
# inch 단위를 입력으로 받아서 cm 단위 변경해주는 프로그램을 만들어볼께요.
# 1 inch = 2.54 cm
# 예시) 입력: 34 inch --> 출력 86.36 cm
# 입력값: OO inch --> 출력값: OO.OO cm
```

```python
# control (cmd) + /
```

```python
one_inch = 2.54 # one_inch <- 2.54
# print(one_inch, type(one_inch))

string_integer = input('정수를 입력해주세요>') # 정수을 입력으로 받음 (근데, 문자열)
# print(string_integer, type(string_integer))

integer = int(string_integer) # 정수형으로 변환
# print(integer, type(integer))

cm = integer * one_inch # 인치를 센치로 계산
print('입력값:', integer, 'inch', '--> 출력값:', cm, 'cm')  # 출력
```

- input() 함수
- input()으로 받은 값은 str
- int() 함수로 str -> int, float -> int
- float() 함수로 str -> float, int -> float
- str() 함수로 int -> str, float -> str
