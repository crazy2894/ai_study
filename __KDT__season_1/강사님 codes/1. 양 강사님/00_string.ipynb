# 자료형과 기본 자료형

- 개발자들이 쉽게 사용할 수 있도록 기능과 역할에 따라 자료를 구분

### 문자열 연산자
- 숫자에는 더하기, 빼기, 곱하기, 나누기 연산을 할 수 있음
- 집합에는 합집합, 교집합, 차집합 등의 연산을 할 수 있음
- 연산자는 숫자나 집합과 같이 같은 특성을 가진 그룹에 종속 됨 (즉, 끼리끼리만 사용 가능)
- 프로그래밍 언어에서 나오는 그룹, 즉 자료형에도 연산자가 각각 존재
- 그럼, 문자열에 적용할 수 있는 연산자는?

#### 문자 선택 연산자(인덱싱): []
- 문자 선택 연산자는 문자열 내부의 문자 하나를 선택하는 연산자
- 대괄호 [] 안에 선택할 문자의 위치를 지정, 이 숫자를 인덱스(index)라고 함
```
프로그래밍 언어에서 사용되는 인덱싱 유형
>> 제로 인덱스 : 숫자를 0부터 셈 (파이썬)
>> 원 인덱스 : 숫자를 1부터 셈
```

|인덱스|0|1|2|3|4|
|-|-|-|-|-|-|
|문자|안|녕|하|세|요|

````python
print('안녕하세요'[0])
````

````python
print('안녕하세요'[1])
print('안녕하세요'[2])
print('안녕하세요'[3])
print('안녕하세요'[4])
````
