# 숫자 자료형

- 정수형 : 소수점이 없는 숫자 (0, 1, 273, -52)
- 실수형 : 소수점이 있는 숫자 (0.0, 52.273, 2.3e10, -1.2)

```python
print(273)
```

```python
print(type(273)) # int -> integer (정수)
```

```python
print(52.273)
```

```python
print(type(52.273)) # float -> floating point (부동 소수점)
```

```python
print(type(0), type(0.0))
```

```python
print(0.52783e3)
```

```python
print(0.52783E3)
```

```python
print(52783E-3)
```

```python
print(52783e-3)
```

```python
print(5.2783e10)
```
