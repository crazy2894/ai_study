## 문자 인코딩 (character Encoding)

- 인코딩 : 정보의 형태나 형식을 변환하는 처리 또는 처리 방식
- 문자 뿐 아니라 사진, 영상 등 다른 데이터 포맷에도 사용됨
- 문자 인코딩 : 글자를 어떤 코드표로 매핑시켜놓은 것
    - ex) '갸' -> 98398492839
- 어떤 데이터를 받고 내 컴퓨터에서 열어봤는데 꿿휄쉙ㅇ!? 처럼 문자가 깨져서 온 적이 있을 것
- 이는 상대방이 보낸 데이터의 인코딩 방식과 내 컴퓨터에서 설정된 인코딩 방식이 달랐기 때문임
- 인코딩이 다르다 : 표현하려는 동일한 문자에 대해 매핑된 코드표의 값(ID)이 다르다는 것을 의미
 
- 여러 인코딩을 사용하는 이유 :데이터를 전송할 때 데이터의 크기를 줄여 최대한 빠르게 저장할 수 있게 하기 위해
    - ex) UTF-8은 8이 붙은 것 처럼 문자를 표현하는데 8bit(1byte)가 소요 / UTF-16은 16bit(2byte)를 차지
- 대표적인 인코딩 종류 : ASCII, CP949, 유니코드, ANSI, UTF-8, UTF-16, EUC-KR
