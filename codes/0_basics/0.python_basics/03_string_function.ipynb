## 문자열 찾기: str.find(), str.rfind()
- str.find() : **왼쪽**부터 찾아서 **처음** 등장하는 위치를 찾음
- str.rfind() : **오른쪽**부터 찾아서 **처음** 등장하는 위치를 찾음

```python
string = "안녕안녕하세요"
print(string)
```

|0|1|2|3|4|5|6|
|-|-|-|-|-|-|-|
|안|녕|안|녕|하|세|요|

```python
print(string.find('안녕'))
```

```python
print(string.rfind('안녕'))
```

```python
print(string.find('이런'))
```
