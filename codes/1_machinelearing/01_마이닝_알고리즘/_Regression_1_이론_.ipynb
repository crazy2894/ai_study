### 상관계수
두변수 간의 선형관계의 강도와 방향을 측정 : 선형이 아닌 상관 관계일 경우에는 다른 상관 관계가 있을 수 있어 측정되지 않을 수 있다.
1 에 가까울 수록 강함 <br>
- 인과관계를 파학하지 못함

### 회귀 계수
독립 변수가 종속 변수에 미치는 크기와 방향을 나타낸다. <br>
상관 계수의 한계로 인해 회귀 계수를 구해야한다. <br>
- 독립 변수와 종속변수와의 인과 관계를 파악

### 상관 과 회귀
- 상관계수는 두변서간의 선형관계, 회귀계수는 독립변수가 종속변수에 미치는 구체적인 영향의 크기, 여려변수 고려가능하다.



### 오차의 종류

SSE(sum of squres) : 실제값과 회귀에서 추정값간의 차이

SST : 전체 평균 간의 차이를 제곱 

SSR : 예측한 추정값과 데이터의 전체 평균 간의 차이를 제곱

```python
import numpy as np
from sklearn.linear_model import LinearRegression

#예제 데이터
X = np.array([2, 3, 4, 5, 6])  # 독립 시간
Y = np.array([5, 7, 9, 11, 13])  # 시험 점수

#상관계수 계산
correlation_matrix = np.corrcoef(X, Y)
correlation_coefficient = correlation_matrix[0, 1]
print("상관계수:", correlation_coefficient)

#선형 회귀 모델 생성 및 훈련
model = LinearRegression()
model.fit(X.reshape(-1, 1), Y)

#회귀계수
regression_coefficient = model.coef_[0]
print("회귀계수:", regression_coefficient)

#결정계수 (R²) 계산
r_squared = model.score(X.reshape(-1, 1), Y)
print("결정계수 (R²):", r_squared)
```
