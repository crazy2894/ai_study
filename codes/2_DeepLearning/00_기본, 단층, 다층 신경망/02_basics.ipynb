**sigmoid**

```python
def sigmoid(t): 
    return 1/(1+np.exp(-t))
```

```python
t = np.linspace(-5, 5, 100)

fig, ax = plt.subplots(figsize = (4,3))

ax.plot(t, sigmoid(t))
ax.set(title = "sigmoid function");
```
