# RNN (Recurrent Neural Networks)

- 시계열, 자연어 (순서가 있는 경우)
- 이전의 정보를 다음의 입력으로 사용하는 구조
- 시퀀스 데이터 처리에 특화

- Sequence 정보가 중요하다
- 음성, 자연어, 시간 순서의 데이터 등에서 사용된다.
- RNN 이라는 하나의 층이 포함 되어있는 모델
- 단어간의 문맥, 관계 정보, 어떤 단어가 언제 사용되었는 지 에 대한 정보 추출이 용이하다

# RNN 과 FNN 모델의 차이

## 정보의 흐름

FNN : 각 입력에 대해 독립적, 시간이나 순서 연관이 없다

CNN : **시간적인 순서(연관성)**, **전체 정보**를 고려한 처리 가능

## 상태의 보존

FNN : Processing 후 상태의 보존이 없다

CNN : **Hidden State(벡터)**를 계속 업데이트 하여 이전 정보를 참조한다.
