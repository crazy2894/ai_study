## 풀링

피쳐 맵의 크기를 줄인다. (좀더 함축적)

**필터와 Activation Map 을 적용하지 않는다**

시각 적으로 확인한 후 어떤 Pooling 을 적용해야 하는지 결정

- Max Pooling : 풀 링 사이즈안의 가장 큰 값을 추출
    - 더 밝은 부분을 추출
    - 관심 있는 물체가 밝은 경우
- Average Polling : 평균을 계산하여 추출
    - 스무싱한 효과
- Global Pooling : 뎁스(차원)은 유지가 되고 각 각 하나의 특정 value 만을 추출하여 임베딩 벡터를 얻을수 있다.

# CNN의 학습 과정

1. 예측 :
    1. CNN 층 : 이미지를 Input 합성 곱, 활성화 함수 등을 이용해 Feature Map 형성 (feature) 추출
    2. FNN 층 : Feature Map 을 Flatten 하여 Softmax 로 라벨에 대한 확률 분포 output
2. 정확도 평가 (손실 함수) : 정답과 예측의 비교
    1. Cross Entropy Loss : 정답(원 핫 인코딩 된) 과 예측의 확률 분포의 차이를 계산
3. 가중치 업데이트 : 역전파를 통하여 가중치를 업데이트 (CNN의 필터와, FNN 의 wieght )
